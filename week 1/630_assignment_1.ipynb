**1.** We will start by checking for covariate balance. Use the statsmodels.stats or scipy stats module to run a (two-sided) t-test to test the following null hypothesis: the mean of *somecol_more* is the same for African-Americans and Whites. Assign the results of the ttest to the variable `ttest1_1`. (1 pt)

**Note:** For this balance test, we could also use a nonparametric test (Chi-squared test for binary variables) that relies on fewer assumptions.

```python
# split the education covariate by race, then compare group means with a two-sided t-test
# black = 1 represents African-Americans; black = 0 represents Whites
somecol_black = data_survey.loc[data_survey["black"] == 1, "somecol_more"]
somecol_white = data_survey.loc[data_survey["black"] == 0, "somecol_more"]

# the test asks whether the two group means are statistically distinguishable
ttest1_1 = stats.ttest_ind(somecol_black, somecol_white)

# display the key values needed for the written interpretation
ttest1_1
```

```python
# Hidden Tests, checking t-statistic and p-value of ttest1_1.
```

**2.** Based on the p-value, do you find evidence that this variable (somecol_more) differs between African-Americans and Whites at the 5% significance level? Explain. (i.e. report BOTH the p-value **AND** the decision rule based on p-value to determine if the difference is significant at the 5% level.) (1 pt)
 
**Note**: This question will be manually graded.

### STUDENT ANSWER

---

The p-value from the t-test is $3.7514 \times 10^{-29}$, which is much smaller than the 5% significance level, $ \alpha = 0.05$. 

The decision rule is: 
- If p-value < 0.05, reject the null hypothesis and conclude that the difference is statistically significant.
- If p-value $ \geq$ 0.05, fail to reject the null hypothesis and conclude that there is not enough evidence of a statistically significant difference.

Since $3.7514 \times 10^{-29} < 0.05$, we reject the null hypothesis. Thus, there is strong statistical evidence that `somecol_more` differes between African-American and White at the 5% significance level.

**3.** Use the statsmodel.stats or scipy stats module to run a (two-sided) t-test to test the following null hypothesis: the mean of years of experience (*yearsexp*) is the same for African-Americans and Whites. Assign the results of the ttest to the variable `ttest1_3`. (1 pt)

```python
# split years of experience by race, then compare the average experience level
# this checks whether experience is balanced between African-Americans and Whites
yearsexp_black = data_survey.loc[data_survey["black"] == 1, "yearsexp"]
yearsexp_white = data_survey.loc[data_survey["black"] == 0, "yearsexp"]

# the two-sided t-test evaluates whether the difference in mean experience is plausibly zero
ttest1_3 = stats.ttest_ind(yearsexp_black, yearsexp_white)

# display the t-statistic and p-value for interpretation
ttest1_3
```

**Note**: This question will be manually graded

### STUDENT ANSWER

---

Based on the t-statistic, there is evidence that `yearsexp` differs between African-Americans and Whites at the 5% significance level.

The decision rule for a two-sided t-test at the 5% level is:
- Reject the null hypothesis if the t-statistic is less than the negative critical value or greater than the positive critical value.
- With a very large degrees of freedom value, df = 10591, the critical values are approx. -1.96 and 1.96.

Thsi rule is: `Reject H0 if t < -96 or t > 1.96`. Since the observed t-statistic is: `-20.6659 < -1.96`, we reject the null hypothesis that Afican-Americans and Whites have the same average `yearsexp`.

**4.** The outcome variable in the data set is the variable *call*, which indicates a callback for an interview. Let’s look at the means of this variable conditional on race.

**4a.** Use the statsmodel.stats or scipy stats module to run a (two-sided) t-test to test the following null hypothesis: the callback rates (means of variable *call*) are the same for resumes with black-sounding names as they are for resumes with white-sounding names. Assign the output to the variable `ttest2_4_1`. (1 pt)



```python
# compare callback rates across the randomized race signal
# because race signal was randomly assigned, this difference is interpretable as causal evidence
call_black = data_random.loc[data_random["black"] == 1, "call"]
call_white = data_random.loc[data_random["black"] == 0, "call"]

# two-sided t-test of equal callback rates between Black-sounding and White-sounding names
ttest2_4_1 = stats.ttest_ind(call_black, call_white)

# display the result for interpretation
ttest2_4_1
```

### STUDENT ANSWER:

At the 5% significance level, the decision rule based on the p-value is:

**5.** Let’s see if the results change when we compare callback rates conditional on gender.

**5a.** Use the statsmodel.stats or scipy stats module to carry out a (two-sided) t-test to test the hypothesis that callback rates are equal for females across races. Assign the result to the variable `ttest2_5_1`. (1 pt)



```python
# restrict to resumes coded as female and compare callback rates across race signal
# this checks whether the racial callback gap appears within the female subgroup
female_black_calls = data_random.loc[
    (data_random["male"] == 0) & (data_random["black"] == 1),
    "call"
]

female_white_calls = data_random.loc[
    (data_random["male"] == 0) & (data_random["black"] == 0),
    "call"
]

# two-sided t-test of equal callback rates for female resumes across race
ttest2_5_1 = stats.ttest_ind(female_black_calls, female_white_calls)

ttest2_5_1
```
