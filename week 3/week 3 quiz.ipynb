### 1. Which of the following is/are true about the instrumental variables (IV) method? Select all that apply.

- [x] Instrumental variables estimates are often imprecise because they are estimated based only on the part of the variation in the treatment that is caused by the instrument (i.e. exogenous variation in the treatment)
- [ ] The instrumental variables approach cannot be used to analyze experiments.
- [x] Instrumental variables approach estimates the average causal effect for the “compliers,” i.e. the local average treatment effect (LATE).
- [ ] The instrumental variables approach addresses selection on observables.

The other statements are false. IV can be used in randomized encouragement experiments, and it addresses endogeneity or selection on unobservables rather than merely selection on observables.

### 2. Which of the following is/are true about IV assumptions? Select all that apply.

- [x] The exclusion restriction cannot be tested.
- [ ] The first stage assumption says that the instrument has a causal effect on the outcome variable.
- [ ] If treatment effects are heterogenous, we can estimate the average treatment effect (ATE) under the “no defiers” assumption.
- [x] The independence assumption says that the instrument is randomly assigned or “as good as randomly assigned.”

The first-stage assumption says the instrument affects the treatment, not the outcome. With heterogeneous effects and no defiers, IV identifies the LATE, not necessarily the overall ATE.

### 3. A bank is interested in knowing whether participation in a week-long management training program makes it more likely that a loan officer is promoted to branch manager. You have data on all the loan officers, their participation in the program, and whether they have been promoted between years three and five of being in their job. Suppose the bank encourages participation in the training program by sending a personal letter from the CEO to selected employees who have been recommended by their supervisor as showing particular potential for a position in management. In the end, 83% of the letter recipients end up attending the program.

#### Would it be useful to use the CEO letter as an instrument for participation in the training program? If not, why?

- [ ] Yes, the CEO letter can be used as an instrument.
- [x] No. CEO letter cannot be used as an instrument because it violates the independence assumption and the exclusion restriction.
- [ ] No. CEO letter cannot be used as an instrument because it violates the first stage and the exclusion restriction.
- [ ] No. CEO letter cannot be used as an instrument because it violates the first stage and independence assumption.

#### Reasoning:

- First stage is satisfied: 83% of recipients attend the program.
- Independence is violated: recipients were selected based on supervisor recommendations rather than randomly.
- Exclusion may be violated: receiving a personalized CEO letter or being identified as having management potential could directly affect promotion, apart from training participation.

### 4. We would like to investigate whether drinking coffee causes stress. An individual’s coffee consumption is likely determined by their personality type, personal motivation, current stress levels, and the price of coffee. Furthermore, personality type and motivation are also likely to affect stress levels. The key idea behind instrumental variables is that although the treatment (coffee consumption) may be endogenous, perhaps there is some portion of its variation that is exogenous. The instrumental variables approach seeks to isolate the exogenous variation in the treatment. Suppose we have data coming from a yoga studio on instructors’ gender, age, and coffee consumption. Additionally, we have access to the price of coffee and the results of a recently conducted motivation level assessment of yoga instructors. Which of these variables is a good candidate for an instrument for coffee consumption?

- [x] Price of coffee 
- [ ] Results of motivation level assessment 

Coffee prices should affect coffee consumption, satisfying the first-stage condition. The motivation assessment is likely related directly to stress, personality, or motivation, so it would violate independence and/or exclusion.

### 5. Which of the below is true about the intention-to-treat effect (ITT)? Select all that apply.

- [ ] We can estimate the ITT with the first stage regression.
- [x] It is the causal effect of being assigned to the treatment on the outcome variable.
- [ ] It can be interpreted as the upper bound of the average treatment effect (ATE).
- [x] We can estimate ITT with the reduced form regression.

The regressions are:

- First stage: assignment → actual treatment
- Reduced form: assignment → outcome

Therefore, the reduced-form coefficient is the ITT.

In particular, in each year from 1970 to 1972, random sequence numbers were randomly assigned to each birth date in cohorts of 19-year olds.

#### IV estimates of the effects of military service on US white men born in 1950:

#### Which of the following is/are true? Select all that apply.

- [x] The first stage estimate suggests that draft-eligible men were 0.16 more likely to have served in the Vietnam era. (Rounded to two decimal places.)
- [x] The reduced form estimate suggests that eligibility reduces earnings by $435.8.

$$

t = \frac {-435.8}{0.159} \approx -2741

$$

Since the approximate F-statistic is greater than 10, the first stage is reasonably strong.

### 9. What are some of the benefits of using the 2-stage least squares (2SLS) method? Select all that apply.

- [ ] It consists of two steps: first stage and reduced form.
- [ ] It guarantees that LATE is equivalent to ATE.
- [x] It allows us to use multiple instruments simultaneously.
- [x] It allows us to measure the precision of IV estimates.

2SLS consists of a first-stage regression and a second-stage regression, not a first stage and reduced form.

It also does not automatically guarantee that LATE equals ATE.
