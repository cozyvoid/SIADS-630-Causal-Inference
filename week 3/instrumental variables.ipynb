### Basic IV Setup 

- Problem: treatment $D$ is affected by unmeasured confounders $U$
- Identify (or create) an *exogenous* variable $Z$ that affects the treatment $D$ but not the outcome $Y$
- $Z$ is called the **intrumental variable** (or "intrument")
- Use the unconfounded variation in the instrument $Z$ to identify treatment effects 



## IV Assumptions 

- Requirements for the instrument $Z$:
    1. Needs to cause changes in the treatment 
    2. "As good as randomly"; cannot influence confounders or vice versa 
    3. Cannot directly influence the outcome 



1. **First stage**: $Z_i$ has a causal effect on $D_i$
    - Rule of thumb: *F*-statistic of a joint test whether all intruments are significantlly different from zero > 10
    - In case of a single instrument: *t*-statistic > $\sqrt 10$
2. **Independence**: $Z_i$ is randomly assigned or "as good  as randomly assigned" 
    - $Z$ is uncorrelated with any possible confounder 
3. **Exclusion restriction**: $Z_i$ does *not* have a direct causal effect on $Y_i$
    - Any effect of $Z$ on $Y$ must go through $D$
    - Technically not testable 

### Finding Good Instruments Can Be Difficult 

- Arbitrary rules created by policies:
    - Angrist and Lavy (1999): Maimonides' rule as an IV for class size (test scores as outcome)
    - Levitt (1997): Being an election year as an IV for police force size (crime as outcome)
- Quasi-random variable in the world:
    - Madestam et al. (2013): Rainfall as an IV for political rally attendance (voting as outcome)
    - Dinkelman (2011): Terrain as an IV for electrification in rural areas (female employment as outcome)

- Randomized encouragement designs 

### Randomized Encouragement Designs 

- Maybe we can re-use an AB test that encouraged users to refer friends 
- Suppose you previously ran an experiment where some users were offered a $5 voucher for referring a friend 

- This experiment serves as an instrument to estimate the causal effect of referrals on app use: 
    - Users refer more friends with $5 referral offer -> first stage :white_check_mark:
    - $5 referral offer was randomly assigned -> independence :white_check_mark:

#### Intention-to-Treat Effect 

- Being assigned to $5 referral offer increased number of articles read: 



- **Intention-to-treat effect (ITT)** is the average causal effect of an *offer* of the treatment (i.e. intrument)
    - ITT is **not** the causal effect of the treatment on the outcome (which is what we want to know)
    - ITT does **not** account for "noncompliance" with treatment assignment (*imperfect compliance*)
