### Basic Idea of IV 



### IV Chain Reaction 

- A is the effect of instrument on treatment (first stage)
- B is the causal effect of interest 
- C is the effect of instrument on outcome via treatment (reduced form)



- While we cannot directly estimate B, we can recover B via estimating A and C as the instrument provides *exogenous* variation in the treatment: 



### IV with Constant Effects 

- Long regression with constant effects: 



- We assume that $E[D_i v_i] = 0$, so if we measured $U_i$, then we would be able to estimate $\pi$.
- But $Cov [\gamma U_i + v_i, D_i] \neq 0$ because $U_i$ is a common cause of $D_i$ and $Y_i$



- If we have an instrument $Z_i$ that satisfies *independence* and the *exclusion restriction*, then: 

$$
Cov (\gamma U_i + v_i, Z_i) = 0
$$

- $Z_i$ must be independent of $U_i$ and it has not correlation with $v_i$ (because $Z_i$ only affects $Y_i$ through $D_i$)

- Now we can identify $\tau$:
$$
Cov (Y_i, Z_i) = Cov (\alpha + \tau D_i + \gamma U_i + v_i, Z_i)
$$
$$
= Cov (\alpha, Z_i) + Cov (\tau D_i, Z_i) + Cov (\gamma U_i + v_i, Z_i)
$$
$$
= 0 + \tau Cov (D_i, Z_i) + 0
$$

### IV Estimator 

- The IV Estimator is the sample analog of: 

$$

\tau = \frac {Cov(Y_i, Z_i)}{Cov (D_i, Z_i)} = \frac {Cov (Y_i, Z_i)/ V(Z_i)} {Cov (D_i, Z_i)/ V(Z_i)}

$$



- Reduced form coefficient: $ Cov(Y_i, Z_i)/V(Z_i)$
- First stange coefficient: $ Cov(D_i, Z_i)/ V(Z_i)$

### IV Estimator with Binary Instrument 

- With a binary instrument, the IV estimator is the sample analog of: 
$$
\tau = \frac {Cov(Y_i, Z_i)} {Cov(D_i, Z_i)} = \frac {E[Y_i|Z_i = 1] - E[Y_i|Z_i = 0]} {E[D_i|Z_i = 1]- E[D_i|Z_i = 0]}
$$

- Inuitively:
$$
\tau = \frac {Effect of instrument on outcome}{Effect of instrument on treatment}
$$
