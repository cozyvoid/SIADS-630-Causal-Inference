### 1. In a study with 1,000 participants, it was found that palm-size correlates with life expectancy, the larger the size of the palm, the shorter is the life expectancy on average. Without a further inspection, one can be tempted to claim that there is a negative causal relationship between the two. It turns out there is a common cause behind this relationship: gender. Women have smaller palms on average compared to men and it is widely known that they have a longer life expectancy. Which of the following explains the causal inference problem between palm-size and life expectancy?

[] Insufficient number of observations
[] Simultaneity
[x] Omitted Variable Bias (selection bias)
[] Reverse causality

**Answer: Omitted Variable Bias (selection bias)**

The apparent relationship between **palm size** and **life expectancy** is misleading because a third variable, **gender**, influences both:

* Gender affects **palm size**: men tend to have larger palms than women.
* Gender affects **life expectancy**: women tend to live longer than men.

So the observed correlation is not necessarily because palm size causes shorter life expectancy. The real issue is that **gender was omitted from the analysis**, creating a biased association between palm size and life expectancy.

### 2. Which of the following are examples for violation of Stable Unit Treatment Value Assumption (SUTVA)? Select all that apply.

[x] In an intervention that offers food assistance to a treatment group, some of the treated subjects receive fewer food stamps than others due to budget constraints.
[] After a behavioral lab experiment is over, treated and control subjects talk to each other about their experiences in the experiment.
[x] A vaccinated subject’s probability of contracting a disease depends on whether others around him/her are vaccinated or not.
[x] An intervention that offers financial assistance to the treatment group is observed by the control group leading to a change in the way those in the control group evaluate their financial conditions.

**Correct answers:**

* **In an intervention that offers food assistance to a treatment group, some treated subjects receive fewer food stamps than others due to budget constraints.**
* **A vaccinated subject's probability of contracting a disease depends on whether others around him/her are vaccinated or not.**
* **An intervention that offers financial assistance to the treatment group is observed by the control group, leading to a change in the way those in the control group evaluate their financial conditions.**

These violate **SUTVA** because SUTVA requires:

1. **No different versions of treatment** — everyone assigned to treatment should receive the same treatment condition.
2. **No interference between units** — one person’s outcome should not depend on another person’s treatment status.

The behavioral lab option is less clearly a SUTVA violation because the subjects talk **after the experiment is over**, so it would not necessarily affect the measured outcomes.

### 3. Consider a randomized control trial with four observations, of which two units were randomly assigned to treatment via complete randomization. We use $ D_i \in {0, 1}$ to denote treatment status (1 for treatment, 0 for control) and $Y_i$ to denote the observed outcome for unit $i$. Based on the provided table, which of the following is true?

| $i$ | $D_i$ | $Y_i$ |
|-----|-------|-------|
| 1 | 1 | 8 |
| 2 | 0 | 2 | 
| 3 | 0 | 4 | 
| 4 | 1 | 6 |


**Correct answer: $(Y_4^1 = 6)$**

Because unit 4 was assigned to treatment:

$
D_4 = 1
$

So its observed outcome equals its treated potential outcome:

$
Y_4 = Y_4^1 = 6
$

The other options are not necessarily true:

* **$(Y_2^1 = 2)$** is false/unknown because unit 2 was in control, so (2) is $(Y_2^0), not (Y_2^1)$.
* **ATT = 4** and **ATE = 4** cannot be known from this table alone because we do not observe both potential outcomes for each unit.
* The value (4) is the **difference in observed group means**:

$
\frac{8+6}{2} - \frac{2+4}{2} = 7 - 3 = 4
$

but that is only the estimated treatment effect, not necessarily the true ATT or ATE.


### 4. We want to investigate the effectiveness of the influenza vaccine. Consider the following case. Amy and Chris are both students at the University of Michigan. Amy regularly exercises, does not smoke, and eats a balanced diet. Chris, on the other hand, does not exercise, smokes a pack of cigarettes every week, and eats mostly fast food. When the University announces that it is time for free flu shots, Amy gets one and Chris does not. In the below table, you can see the actual and potential health outcomes, as well as treatment assignments of Amy and Chris. Bolded entries are the hypothetical results since we can only observe one of the potential outcomes for each.

| Subject | Treatment | Actual Outcome | Potential Outcome | Potential Outcome | 
|----|----|----|----|----|
| $i$ | $D_i$ | $Y_i$ | $Y_i^1$ | $Y_i^0$ |
| Amy | 1 | 7 | 7 | 5 | 
| Chris | 0 | 2 2| 5 | 2 | 

The difference in actual outcomes between Amy and Chris is equal to $(Y_A | D_A = 1) - (Y_C | D_C = 0) = Y_A^1 - Y_C^0 = 7 - 2 = 5$

How much of that is due to selection bias?

[] 2
[x] 3

**Correct answer: 3**

The observed difference is:

$
Y_A^1 - Y_C^0 = 7 - 2 = 5
$

To find **selection bias**, compare what Amy and Chris’s outcomes would have been **under the same untreated condition**:

$
Y_A^0 - Y_C^0 = 5 - 2 = 3
$

So **3 points** of the observed difference are due to selection bias.

The remaining part is Amy’s treatment effect:

$
Y_A^1 - Y_A^0 = 7 - 5 = 2
$

So:

$
5 = 2 \text{ treatment effect} + 3 \text{ selection bias}
$


### 5. Which of the following are true statements? Select all that apply.

[] Every (point) estimate has an associated standard error.
[] Results are called internally valid if they are unbiased for the full population.
[] When we randomly assign individuals (from a large enough sample) to a treatment or a control condition, the two groups have the same expected potential outcomes.
[] Random sampling ensures that a sample statistic (e.g. sample mean) is always equal to the corresponding population parameter

**Correct answers:**

* **Every (point) estimate has an associated standard error.**
* **When we randomly assign individuals from a large enough sample to a treatment or control condition, the two groups have the same expected potential outcomes.**

Why:

The first is true because a point estimate, such as a sample mean or estimated treatment effect, has sampling uncertainty, which is summarized by a **standard error**.

The third is true because **random assignment** makes treatment status independent of potential outcomes in expectation. With a large enough sample, the treatment and control groups should be comparable on average.

The other two are false:

* **Internal validity** means the study gives an unbiased estimate of the causal effect for the study/sample context, not necessarily the full population.
* **Random sampling** does not guarantee the sample statistic exactly equals the population parameter; it only makes the sample statistic unbiased in expectation.

At the 5% significance level, we can reject the null hypothesis if the t-statistic is larger than roughly 2 (two-sided t-test). Based on this, which of the following is true? ### 7. Which of the following statement(s) is/are true about standard errors? Select all that apply.

[x] Standard error is a measure of the variability of (point) estimates.
[] Standard error increases with the size of the sample.
[x] Standard errors can be used for hypothesis testing to help us answer questions like “Is there a significant difference in means between the treatment group and the control group?”
[] Standard error is a measure of the variability of individual values

**Correct answers:**

* **Standard error is a measure of the variability of point estimates.**
* **Standard error can be used for hypothesis testing to help us answer questions like “Is there a significant difference in means between the treatment group and the control group?”**

The false options:

* **“Standard error increases with the size of the sample”** is false. Standard error usually **decreases** as sample size increases.
* **“Standard error is a measure of the variability of individual values”** is false. That describes the **standard deviation**, not the standard error.


### 8. Why are randomized experiments considered the “gold standard”? Select all that apply.

[] They are easily applicable to almost every context.
[x] Randomized experiments have the highest internal validity (provided the sample is large enough).
[x] They guarantee that the treatment assignment is a fair process, i.e. everyone can get the treatment.
[x] They eliminate selection bias on observables and unobservables.

**Correct answers:**

* **Randomized experiments have high internal validity, provided the sample is large enough.**
* **They guarantee that the treatment assignment is a fair process, i.e., everyone can get the treatment.**
* **They eliminate selection bias on observables and unobservables.**

The false option is:

* **They are readily applicable to almost every context.**

Randomized experiments are powerful because random assignment makes treatment and control groups comparable **in expectation**, including both observed and unobserved characteristics. But they are not always feasible because of ethical, logistical, legal, or cost constraints.


### 9. What is the fundamental problem of causal inference?

[] We cannot always run randomized experiments.
[] We cannot always use samples to infer the properties of a population.
[] We cannot always find enough observations to investigate a causal relationship.
[x] We cannot observe the counterfactual.

**Correct answer: We cannot observe the counterfactual.**

The **fundamental problem of causal inference** is that for the same person/unit, we can observe only one potential outcome:

$
Y_i^1 \quad \text{or} \quad Y_i^0
$

but not both.

For example, Amy either gets the flu shot or does not get the flu shot. We can observe what happens under the condition she actually experienced, but we cannot directly observe what would have happened to the same Amy under the opposite condition. That missing outcome is the **counterfactual**.


### 10. Which of the below is true about the ATE (average treatment effect) and the ATT (average treatment effect on the treated)?

[] ATT is the average treatment effect for people who agreed to participate in the study.
[x] ATE is the average change in outcomes if everybody in the population were to go from untreated to treated.
[] Based on the ATE, we can determine if the majority of the people have a positive or negative treatment effect.
[] ATE is not a quantity we can calculate, we can only estimate it; whereas ATT is a quantity we can calculate.

**Correct answer:**

**ATE is the average change in outcomes if everybody in the population were to go from untreated to treated.**

Formally:

$
ATE = E[Y^1 - Y^0]
$

It describes the average causal effect across the full population.

The others are incorrect:

* **ATT** is the average treatment effect for those who actually received treatment, not simply those who agreed to participate.
* **ATE** tells us the average effect, but not whether the majority of individuals had positive or negative effects.
* Both **ATE** and **ATT** usually cannot be directly calculated because we cannot observe every unit’s counterfactual outcome; both must generally be estimated.
