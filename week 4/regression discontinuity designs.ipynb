# Regression Discontinuity Designs 

---

### Example - The Effect of MADS on Earnings

- Imagine you want to estimate the effect of passing the MADS causal inference course on earned income 

- randomly assigning some students to the "passing" treatment and failing others is unethical 
- controlled regression might be biased of unobservaed confounders 
- passing cutoff creates a natural experiment 





### Regression Discontinuity Designs 







**Regression discontinuity (RD)** designs exploit "arbitary" cutoffs embedded in rules 



### Basic RD Setup 

- $X_i$ is the **running variable**
- Treatment assignment is determined by a cutoff in $X_i$
- Relationship between potential outcomes and running variable has to be "smooth" at the cutoff 
- Changes in the outcome around the threshold can be interpreted as causal effects
