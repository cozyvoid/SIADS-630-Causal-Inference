### 1. Which of the following is/are true about regression discontinuity (RD) designs? Select all that apply.

- [ ] The validity of RD does not require the cutoff rule to be precise.
- [x] We need the conditional expectation functions of the potential outcomes $(E[Y_I^0|X_i = x]$ and $E[Y_i^1|X_i = x])$ to be continuous or “smooth” at the cutoff.
- [x] RD design requires a lot of data around the threshold.
- [ ] For units far away from the threshold, it is “as good as random” whether they are below or above the cutoff (i.e. whether they are treated or not).

RD identifies a local treatment effect near the cutoff, so observations close to the threshold are essential. The “as good as random” comparison applies only near the cutoff, not far away.

### 2. The key to identification in RD designs is the “continuity assumption.” That is, the expected potential outcomes remain continuous or “smooth” through the cutoff. Which of the cases below is an example of a violation of the continuity assumption?

- [ ] We would like to estimate the effect of getting a distinguished scholarship on students’ future earnings. The scholarship is allocated to students who get more than half of the allocating committee’s votes. The committee has long been known to use blind evaluations in order not to favor any applicants.
- [x] We would like to estimate the effect of harsher punishments on the likelihood of committing crime. When a person turns 18, he/she not only faces more severe penalties for crime but also graduates from high school, the last leg of compulsory schooling. Being occupied with school is known to reduce crime rates for the age group of 15 – 20 years old.

Turning 18 changes both the severity of punishment and high-school attendance. Because graduation from compulsory schooling could independently affect crime, the outcome may change discontinuously at age 18 for reasons other than punishment.

### 3. We discussed reducing the bandwidth (width of the window around the cutoff) and restricting our estimation to the observations close to the threshold. Which of the following is/are true about reducing the bandwidth? Select all that apply.

- [ ] Decreasing the bandwidth leads to a higher risk of bias.
- [ ] Increasing the bandwidth is one way we can deal with nonlinearities in the data.
- [x] In practice, the choice of the bandwidth typically requires a judgment call.
- [x] Increasing the bandwidth often leads to higher precision of the estimate.

A wider bandwidth includes more observations, which generally lowers variance and increases precision. However, it can increase bias because observations farther from the cutoff may be less comparable.

### 4. Which of the contexts below is suitable for a fuzzy regression discontinuity (RD) design?

- [ ] We would like to investigate the effect of an extra Yelp star on restaurants’ revenues. Users assign a rating from one to five stars in whole star increments. Yelp then rounds off the average ratings to the nearest star, creating cutoffs for average ratings right in the middle of each star. (e.g. an average rating of 3.6 stars would be rounded off to 4 stars)
- [x] We would like to investigate the effect of receiving food stamps on nutrient intake. Food stamp eligibility is given to all households below a certain income, but not all households below the income threshold get food stamps.

Income eligibility changes at the threshold, but not every eligible household receives food stamps. Therefore, crossing the cutoff changes the probability of treatment rather than perfectly determining treatment.

The Yelp example is closer to a sharp RD because the displayed rating is mechanically determined by the rounding rule.

### 5. If individuals have full control over the running variable, that would threaten the identification of RD designs. What are some of the ways we can look for signs of manipulation? Select all that apply.

### 7. Which of the following is/are true about the parallel trends assumption? Select all that apply.

- [x] We cannot test the parallel trends assumption.
- [ ] It requires that the absolute levels of confounding variables in the pre-treatment period are the same in the treatment and control groups.
- [x] We need data from more than just one pre-treatment period to inspect parallel trends assumption.
- [ ] It requires that no unobservable confounder changes in either the treatment group or the control group.

Multiple pre-treatment periods allow us to examine whether the groups followed similar trends before treatment. However, the post-treatment counterfactual trend cannot be directly observed, so the assumption cannot be conclusively tested.

The groups do not need to have the same outcome levels or identical confounder levels.

### 9. Many differences-in-differences applications use data from many periods (not just 1 pre and 1 post-period). One issue with panel data is that there might be serial correlation in the dependent variable, which would lead to downward biased standard errors when we use conventional methods. Which of the following is/are true about serial correlation? Select all that apply.

- [x] We can block bootstrap (replicate the correlation by resampling blocks of data) the standard errors. (Remember: Bootstrapping is a resampling procedure that uses data from one sample to generate a sampling distribution by repeatedly taking random samples from the known sample, with replacement.)
- [ ] We can collapse the time series into one period for each group.
- [x] If there is serial correlation, we may reject the null hypothesis of no treatment effect too frequently if we use conventional methods for the calculation of standard errors.
- [ ] We can cluster standard errors at the group level, regardless of the number of groups/clusters.

Serial correlation commonly causes conventional standard errors to be underestimated, increasing false-positive findings.

Clustering is not automatically reliable with very few groups or clusters.

### 10. When the intervention is not as good as random, the main identifying assumption of the differences-in-differences strategy (i.e. parallel trends) might be violated. What are some of the methods we can try if we think the parallel trends assumption is violated? Select all that apply.

- [ ] We can drop the data points that we suspect are causing the violation of the parallel trends assumption and apply the differences-in-differences strategy to the subsample.
- [x] We can perform a falsification test by replacing the outcome variable with another outcome that is not supposed to be affected by the intervention.
- [x] If we have more than two periods, we can add unit-specific linear time trends to the regression model to check for whether the parallel trends assumption is violated.
- [ ] We can add fixed covariates (e.g. gender, location etc.) in the regression model in a linear, additive way.

Simply adding fixed, time-invariant covariates such as gender or location in an additive form generally does not correct nonparallel trends. They would need to be incorporated in a way that allows their effects to vary over time.
